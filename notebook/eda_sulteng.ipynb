```python
# Load data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/processed/sulteng_clean.csv")
df.head()
```

```python
# Statistika Deskriptif
df.describe()
```

```python
# Distribusi tiap fitur
num_cols = df.drop(columns=["kabupaten"]).columns

for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribusi {col}')
    plt.show()
```

```python
# box plot deteksi outlier
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[num_cols])
plt.xticks(rotation=45)
plt.title("Boxplot Semua Indikator")
plt.show()
```

```python
# korelasi antar fitur
plt.figure(figsize=(8,6))
sns.heatmap(df[num_cols].corr(), annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Korelasi Antar Indikator")
plt.show()
```

```python
# Ranking Kabupaten per Indikator
for col in num_cols:
    display(df[["kabupaten", col]].sort_values(by=col, ascending=False))
```
